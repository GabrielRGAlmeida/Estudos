<center><span style='font-weight: bold; font-size: 30px'>Aplicando funções</span></center>
<br><br> Podemos criar funções que são aplicadas em todos os valores de uma linha ou de uma coluna. Para fazer isso, definimos uma função qualquer e aplicamos ela dentro do método <span style='font-style: italic'>dataframe.apply(function, axis=) </span>. Se queremos aplicar a função em todas as linhas, usamos <span style='font-style: italic'>axis=1</span> e criamos uma nova coluna para o dataframe <br><center><span style='font-style: italic'>dataframe['new_column'] = dataframe.apply(function, axis=1)</span> </center> Se queremos somar os valores contidos em cada coluna, criamos uma nova linha que conterá o resultado utilizando o apply <br> 
<center> <span style='font-style: italic'>dataframe.loc['new_column'] = dataframe.apply(function, axis=0)</span> </center>


```python
def soma(linha):
    return linha.sum()


test = {
'refrigerantes': [1, 2, 1, 3, 2],
'pipocas': [2, 3, 1, 2, 1],
'água mineral': [1, 0, 2, 0, 1],
'pizzas': [1, 0, 0, 2, 1],
'cachorros quentes':[1, 2, 1, 4, 1]
}
pessoas = [f'pessoa {i}'  for i in range(1,6)] 
tabela = pd.DataFrame(test, index=pessoas)

if 'Total linha' in tabela[:]:
    del tabela['Total linha']

tabela['Total linha'] = tabela.apply(soma, axis=1)
if 'resultado coluna' in tabela.index:
    tabela.drop('resultado coluna', axis=0, inplace=True)
tabela.loc['resultado coluna'] = tabela.apply(soma, axis=0)

tabela
```

```python
tabela['media'] = tabela[['refrigerantes','pipocas','água mineral','pizzas','cachorros quentes']].apply(lambda x: x.mean(), axis=1)
tabela
```

A função map age parecido com o apply, mas ele lê somente uma series e não um eixo. Usamos o método map <span style='font-style: italic'>Series.map(function)</span>. Note também que a função não contém o parâmetro  <span style='font-style: italic'>inplace=True</span>. Para salvar a função, basta criar uma nova variável que salva essa aplicação.

```python
tabela.columns.map(lambda x: x.upper())
```

<center><span style='font-weight: bold; font-size: 30px'>Agrupamentos</span></center>
<br><br> O agrupamento funciona usando um mapper ou uma Series de um DataFrame pegando um objeto, fatiando-o e aplicando uma função para depois agrupar o resultado da função em um novo DataFrame mas do tipo DataFrameGroupBy. Usamos o método <span style='font-style: italic'>DataFrame.groupby('column')</span> que retorna um objeto do tipo dicionário composto por chaves que serão colunas e atribuímos esse método a uma variável qualquer que aqui chamaremos de '<span style='font-style: italic'>group</span>'.   Para mostrar o agrupamento em dicionários, usamos o atributo <span style='font-style: italic'>group.groups</span>. <br>
Outro método para checar os índices, basta usar o atributo <span style='font-style: italic'>group.indices</span> <br>
Para analisar esses grupos em termos de tabela, basta usar o método <span style='font-style: italic'>group.get_group('unique_value')</span> onde esse unique value é um dos valores chaves do dicionário group.

```python
regioes = gasdfpre.groupby('REGIAO')

regioes.groups
```

```python
regioes.indices
```

```python
regioes.get_group('SUL')
```

Esses grupos criados podem ser vistos em termos de estatísticas descritivas usando o método <span style='font-style: italic'>DataFrame.groupby('column').describe()</span>.
<br><br>Além do describe(), se quisermos ver isoladamente alguma (ou algumas) das medidas estatísticas, podemos usar os seguintes métodos <span style='font-style: italic'>.mean(), .min(), .max(), .std()</span>. <br>
Caso o group contenha colunas com valores que não são numéricos, podemos criar uma lista que contenha as colunas que possuem apenas valores numéricos e usamos ela dentro do group através de lista usando o comando <span style='font-style: italic'>group[lista]</span>

```python
regioes.describe()
```

```python
regioes.get_group('NORTE').max()
```
