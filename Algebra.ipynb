   <h3 style='font-weight: bold; font-size: 15px'> Def. 1.1: Conjuntos</h3>
         Dizemos que $A$ é um conjunto (ou coleção) se há uma regra de pertinência $\in$ com algum objeto arbitrário x.<br>
         Denotamos o conjunto como $A = \{x\;|\; x \in A\}$.<br><br><br>
    <h3 style='font-weight: bold; font-size: 15px'> Def. 1.2: Subconjuntos</h3>
             Sejam $A = \{x\; | \; x \in A\}$ e $B = \{y\; | \; y \in B\}$ dois conjuntos. Dizemos que há uma relação de inclusão $A \subset B$ se todos os elementos de $A$ estiverem em $B$, então dizemos que $A$ é um subconjunto de $B$, ou $A$ está contido em $B$ 
         $$A \subset B$$ ou
         $B$ contém $A$ $$B \supset A$$
         <br><br><br>
     <h3 style='font-weight: bold; font-size: 15px'> Def. 1.3: Relação de igualdade entre conjuntos</h3>
     Chamamos de igualdade entre conjuntos se $A \subset B$ e $B \subset A$. Denotamos isso por $A = B$<br><br><br>
     <h4 style='font-weight: bold; font-size: 12px'> Propriedade 1.1: </h4>
     A relação de inclusão $A \subset B$, $B \subset C$ possui as seguintes propriedades:
     <br>  i) <span style='font-style: italic'>Reflexiva</span>: $A \subset A$ <br> 
           ii)<span style='font-style: italic'>Anti-simétrica</span>: $A \subset B \; , \; B \subset A \; \Rightarrow \; A = B$ <br> 
           iii)<span style='font-style: italic'>Transitiva</span>: $A \subset B \; , B \subset C  \; \Rightarrow A \subset C$ <br>
           <br>  <h5 style='font-weight: bold; font-size: 12px; font-style: italic'> Dem. 1.1:</h5> <br>
           i) A demonstração é direta pela definição 1.2; <br>
           ii) A demonstração é direta pela definição 1.3; <br>
           iii) $x \in A \Rightarrow x \in B, \;\;\; x \in B \Rightarrow x \in C  \;\; \therefore A \subset C$ 
     <h3 style='font-weight: bold; font-size: 15px'> Def. 1.4: Conjunto Vazio</h3> 
           O conjunto vazio denotado por $\emptyset = \{x \; | \; x \notin \emptyset\}$ é o conjunto que não possui nenhum elemento.<br><br>
           <h4 style='font-weight: bold; font-size: 12px'> Propriedade 1.2:</h4> 
           Para todo conjunto $A$ qualquer, tem-se que $\emptyset \subset A$.
           <br>  <h5 style='font-weight: bold; font-size: 12px; font-style: italic'> Dem. 1.2:</h5> 
           Suponha que $ \emptyset \not\subset A $ então $\exists x \; : x \in \emptyset $ e $ x \not\in A$, mas $x \in \emptyset$ é uma contradição de acordo com a definição 1.4. <br>
           <h3 style='font-weight: bold; font-size: 15px'> Def. 1.5: Operação entre conjuntos - União</h3> 
           A união entre $A$ e $B$ é operação $A \cup B = \{x \; | \; x \in A $ <b>OU</b> $ x \in B\}$ que resulta no conjunto composto por todos elementos de $A$ e de $B$.<br>
           <h3 style='font-weight: bold; font-size: 15px'> Def. 1.6: Operação entre conjuntos - Intersecção</h3> 
           A intersecção entre $A$ e $B$ é operação $A \cap B = \{x \; | \; x \in A $ <b>E</b> $ x \in B\}$ que resulta no conjunto composto por todos elementos <b>comuns</b> entre $A$ e de $B$.<br>
           <h4 style='font-weight: bold; font-size: 12px'> Propriedade 1.3:</h4>
           Dados os conjuntos $A, B, C$, as seguintes propriedades seguem das definições 1.4 e 1.5:<br>
           Propriedades da União:<br>
           i)$\cup 1$: $A \cup \emptyset = A$ <br> 
           ii)$\cup 2$: $A \cup A = A$ <br> 
           iii)$\cup 3$: $ A \cup B = B \cup A $ <br>
           iv)$\cup 4$: $ A \cup (B \cup C) = (A \cup B) \cup C $<br>
           v)$\cup 5$: $ A \cup B = A \iff B \subset A$ <br>
           vi)$\cup 6$: $A \subset B, \;\; A_1 \subset B_1 \Rightarrow A \cup A_1 \subset B \cup B_1 $ <br>
           vii)$\cup 7$: $ A \cup (B\cap C) = (A \cup B) \cap (A \cup C)$ <br>
           Propriedades da Intersecção: <br>
           i)$\cap 1$: $A \cap \emptyset = \emptyset$ <br> 
           ii)$\cap 2$: $A \cap A = A$ <br> 
           iii)$\cap 3$: $ A \cap B = B \cap A $ <br>
           iv)$\cap 4$: $ A \cap (B \cap C) = (A \cap B) \cap C $<br>
           v)$\cap 5$: $ A \cap B = A \iff A \subset B$ <br>
           vi)$\cap 6$: $A \subset B, \;\; A_1 \subset B_1 \Rightarrow A \cap A_1 \subset B \cap B_1 $ <br>
           vii)$\cap 7$: $ A \cap (B\cup C) = (A \cap B) \cup (A \cap C)$ <br>
           <h3 style='font-weight: bold; font-size: 15px'> Def. 1.7: Operação entre conjuntos - Diferença</h3> 
            Sejam $A$ e $B$ dois conjuntos. A diferença entre $A$ e $B$ é o conjunto $$A-B = \{x \; | \; x \in A, \; x \not\in B\}$$ Um caso especial se dá quando $ B \subset A$ podemos chamar a diferença $A-B =\complement_A B$  como complementar de $B$ em relação a $A$
           <h3 style='font-weight: bold; font-size: 15px'> Def. 1.8: Conjunto Complementar:</h3>

   2 - Funções
</h1>
<h3 style='font-weight: bold; font-size: 15px'> Def. 2.1: Funções:</h3>
            Uma função $f: A \rightarrow B$ é uma aplicação lógica em que relaciona dois conjuntos, o domínio $A$ e o contra-domínio (ou imagem) $B$, obedecendo as seguintes regras: <br>
            i) $\forall x \in A \; \exists y \in B$ <br>
            ii)$f(x) = y_1  \;, \; f(x) = y_2 \; \Rightarrow y_1 = y_2$<br>
            A função $f(x)$ é a aplicação lógica num dado elemento $x \in A$ que corresponde a um valor $y \in B$. Representamos isso por $x \mapsto f(x)$.
<h3 style='font-weight: bold; font-size: 15px'> Def. 2.2: Igualdade entre funções:</h3>
As funções $f: A \rightarrow B$ e $g: C \rightarrow D$ são iguais se, e somente se, $A = C$, $B = D$ e $f(x) = g(x) \; \; \forall x \in A$
<h3 style='font-weight: bold; font-size: 15px'> Def. 2.3: Função Injetiva:</h3>
Uma função $f: A \rightarrow B$ é dita injetiva (ou injetora) quando para quaisquer $x, y \in A, \;\; f(x) = f(y) \Rightarrow x = y$
<h3 style='font-weight: bold; font-size: 15px'> Def. 2.4: Função sobrejetiva:</h3>
Uma função $f: A \rightarrow B$ é dita sobrejetiva (ou sobrejetora) quando $\forall f(x) \in B \; \exists x \in A$
<h3 style='font-weight: bold; font-size: 15px'> Def. 2.5: Função Bijetiva:</h3>
Uma função $f: A \rightarrow B$ é dita bijetiva (ou bijetora) quando ela é injetiva e sobrejetiva ao mesmo tempo. 
<h3 style='font-weight: bold; font-size: 15px'> Def. 2.6: Imagem de $f$:</h3>
Seja $f: A \rightarrow B$ uma função. A imagem de $f$ é o conjunto $$Im(f) = \{f(x) \; | \; x \in A\}$$
<h3 style='font-weight: bold; font-size: 15px'> Def. 2.7: Imagem de X pela função $f$:</h3>
Seja $f: A \rightarrow B$ uma função e $X \subset A$. A imagem de X pela função $f$é o conjunto $$Im(f(X)) = \{f(x) \; | \; x \in X\}$$
