<center><b><span style='font-size: 26px'>Analisando dados</span></center></b><br><br>

Podemos correlacionar os elementos da matriz colunasXcolunas usando diferentes métodos de correlação: <b>Pearson</b>, <b>Kendall</b> e <b>Spearman</b><br><br>
<ul>
<li><b>Pearson</b>: &nbsp; Mede a relação estatística entre duas variáveis contínuas. Se a associação entre elementos tiver uma fraca correlação, o coeficiente de Pearson terá valor próximo de 0. O coeficiente varia entre [-1,1] e valores próximos das extremidades indicam forte correlação linear. Para correlações não-lineares, o coeficiente de Person pode ter valores próximos das extremidades, mas mesmo assim não indicam uma forte relação já que o coeficiente só tem capacidade de medir as correlações lineares.
    $$r = \frac{\Sigma[(X_i - \bar{X})(Y_i - \bar{Y}) ]}{\sqrt{\Sigma(X_i - \bar{X})^2 \Sigma (Y_i - \bar{Y} )^2}} $$
    <br>
    Onde $X_i$ e $Y_i$ representam os valores individuais das duas variáveis, $\bar{X}$ e $\bar{Y}$ representam as médias das variáveis.<br><br><br>
    
<li><b>Kendall</b>: &nbsp; Medida estatística que avalia a associação entre duas variáveis. Considerando um conujunto de observações das variáveis aleatórias conjuntas $X$ e $Y$ tal que $X$ e $Y$ sejam únicos para qualquer par de observações $X_i$ e $X_j$ com $i \neq j$. Dizemos que:
    <ol>
        <li>Par de observações concordantes: <br><br>
            $ X_i < X_j \rightarrow Y_i < Y_j$ &nbsp; &nbsp; &nbsp; &nbsp; ou &nbsp; &nbsp; &nbsp;  $X_i > X_j \rightarrow Y_i > Y_j$<br><br>
        <li>Par de observações discordantes:<br>
            $ X_i < X_j \rightarrow Y_i > Y_j$ &nbsp; &nbsp; &nbsp; &nbsp; ou &nbsp; &nbsp; &nbsp;  $X_i > X_j \rightarrow Y_i < Y_j$<br><br><br>
    </ol>
<li><b>Spearman</b>: &nbsp;Medida estatística que mede a relação monótonas (Lineares e não-lineares). Ela avalia se, o aumento ou dimuição de uma variável também aumenta ou diminui a outra variável.
    $$ r_s = \rho_{rg_X, rg_Y} = \frac{cov(rg_X, rg_Y)}{\sigma_{ rg_X }\sigma_{rg_Y}}$$
     <ul>
        <li>$\rho$ é o coeficiente de correlação Pearson, mas aplicando às variáveis em postos
        <li>$cov(rg_X, rg_Y)$ é a covariância das variáveis em postos
        <li>$\sigma_{ rg_X }$ e $\sigma_{rg_Y}$ são os desvios padrão das variáveis em postos
     </ul>
    
</ul>


<br>Usamos o método <i style='font-weight: 500'>DataFrame.corr(method=, min_periods=, numeric_only)</i> para analisar as correlações. <br>
<ul> 
    <li>parâmetro <i style='font-weight: 500'>method=</i> pode receber os valores <i style='font-weight: 500'>'pearson', 'kendall'</i> ou <i style='font-weight: 500'>spearman</i>
    <li>parâmetro <i style='font-weight: 500'>min_periods=</i> recebe um número intde observações não nulas necessárias para calcular uma correlação válida entre um par de colunas. Por exemplo, se especificarmos o valor 10, então devemos ter 10 valores não nulos em cada coluna para poder calcular a correlação. Caso não haja o valor necessário, então retorná um valor NaN indicando que não pode calcular a correlação entre as colunas.
    <li>parâmetro <i style='font-weight: 500'>numeric_only</i> recebe valor booleano (verdadeiro ou falso) especificando se deve ignorar as variáveis categóricas.
</ul>

<b><span style='font-size: 20px'>Categorical Variables</span></b><br><br>
Variáveis categóricas são variáveis que descrevem alguma característica do dataset. Elas podem ser tanto object quanto int64. Uma ótima escolha para analisar variáveis categóricas é usando o método <i style='font-weight: 500'>boxplot</i> da biblioteca seaborn. Este método gera um gráfico formado pelos valores categóricos da coluna x em relação a variável dependente y (que é a coluna 'price'). Para cada valor de x do gráfico é atribuido um diagrama caixa com as seguintes características:
<ul>
    <li><b>Intervalo Interquartil - IQR (caixa)</b>: Conjunto composto por valores entre o primeiro quartil $Q_1 = 25º$ percentil e terceiro quartil $Q_3 =75º$ percentil;
    <li><b>whisker (pavio)</b>: Conjunto de valores $x \in [Q_1 -1.5 \times IQR, Q_3 + 1.5\times IQR]$;
    <li><b>Outliers (pontos) </b>: São anomalias que estão fora do intervalo $[Q_1 -1.5 \times IQR, Q_3 + 1.5\times IQR]$.
</ul><br><br>
Lembre-se que a mediana é o valor central de um conjunto de dados ordenados, ou seja é o valor do dado que divide os dados em 50% dados são maiores que eles e 50% são menores. Nessa lógica, o primeiro quartil $Q_1$ é o valor do dado que se encontram $\frac{1}{4}$ dos dados e o terceiro quartil $Q_3$ é o onde se encontram $\frac{3}{4}$ dos dados. Assim
$$IQR = Q_3 - Q_1 $$

<b><span style='font-size: 26px'>Correlação e Casualidade</span></b><br><br>

Correlação: a medida estatística que descreve o grau de relação ente duas variáveis.<br>
Casualidade: relação mais forte onde uma mudança em uma variável implica mudança na outra variável.<br>

A correlação não implica casualidade, já que por exemplo pode haver uma correlação entre sorvetes e pessoas na pria. Mas isso não implica que a venda de sorvetes vá aumentar o número de pessoas na praia. Determinar a correlação é muito mais simples que determinar a casualidade já que a casualidade demonstra como uma variável afeta a outra.
<br><br>

<b><span style='font-size: 19px'>P-value</span></b><br><br>

O P-value é uma medida estatística que ajuda a determinar a significância de um resultado. Ele é uma medida probabilística que determina se a correlação entre duas variáveis $X$ e $Y$ é estatisticamente relevante. <br>
Dizemos que um nível de significância de 0.05 equivale a 95% de confidência em que a variável $X$ está relacionada com $Y$, logo:
<ul>
    <li>P-value $< 0.001 \;\; \rightarrow \;\;$ forte evidência <br><br>
    <li>P-value $< 0.05 \;\; \rightarrow \;\;$ moderada evidência <br><br>
    <li>P-value $< 0.1 \;\; \rightarrow \;\;$  fraca evidência <br><br>
    <li>P-value $> 0.1 \;\; \rightarrow \;\;$  não há evidência de correlação <br><br>
</ul>

<br><br> 
Para calcular o coeficiente de Pearson e o P-value, usamos o método <i style='font-weight: 500'>spy.stats.pearsonr($X$, $Y$)</i>. 

```python
x = car_clean[['wheel-base', 'horsepower', 'length', 'width', 'curb-weight', 'engine-size', 'bore', 'city-L/100km', 'highway-L/100km' ]]

for column in x:
    coef, pval = scipy.stats.pearsonr(x[column], car_clean['price'])
    print(f'O coeficiente de Pearson da variável {column} é {coef} e o P-value é {pval}')
```

    <b><span style='font-size: 26px'>Analysis of Variance (ANOVA)</span></b><br><br>    
    A Análise da Variância é um método estatístico que mostra se há uma diferença significante entre as médias de dois ou mais grupos. Ela nos retorna dois parâmetros:
    <br>    
    <ul>
    <li> <b>F-test</b>: Esse parâmetro assume que as médias entre os grupos são iguais e calcula quanto a média atual se desvia do valor assumido. 
    <li> <b>P-Value</b>: Medida estatística que mede a significância de um resultado.
    </ul>
    <br><br>
    Ela consiste na decomposição da variância total em duas componentes: a variância entre os grupos e a variância dentro dos grupos. Para calcular o <b>F-test</b> e <b>P-value</b> ao mesmo tempo, usamos o comando
    <br><i style='font-weight: 500'>x, &nbsp; y =  scipy.stats.f_oneway(group1, group2,...)</i> onde x é o <b>F-test</b> e y é o <b>P-value</b>
    <br><br><br>    
    O <b>F-test</b> é calculado da seguinte forma:
    <div class="minhacaixa">
        <ol>
            <li> <b>Defina as hipóteses</b>:
                <ul>
                    <li> Hipótese nula (H0): As médias dos grupos são iguais
                    <li> Hipótese alternativa (H1): Pelo menos uma das médias dos grupos é diferente
                        <br><br><br>
                </ul>
            <li> <b>Calcule a Soma dos Quadrados (SS)</b>:
                 <ul>
                    <li> Soma dos Quadrados Total (SST): É a soma dos quadrados das diferenças entre cada observação e a média geral. Seja $y_i$​ a i-ésima observação e $\bar{y}$ a média das observações, a SST é dada por:
                    $$SST = \sum_{i=1}^{n}(y_j - \bar{y})^{2} $$<br><br>
                    <li> Soma dos Quadrados dos Grupos (SSG): É a soma dos quadrados das diferenças entre a média de cada grupo e a média geral, ponderada pelo tamanho do grupo. Seja $\bar{y}_j$​ a média do j-ésimo grupo e $n_j$​ o número de observações no j-ésimo grupo, a SSG é dada por:
                        $$SSG = \sum_{i=1}^{k}n_j(\bar{y}_j - \bar{y})^{2}  $$<br><br>
                    <li> Soma dos Quadrados dos Resíduos (SSR):
                        É a soma dos quadrados das diferenças entre cada observação e a média do seu grupo. Seja $y_{ij}$​ a i-ésima observação do j-ésimo grupo, a SSR é dada por: 
                        $$SSR = \sum_{j=1}^{k}\sum_{i=1}^{n_j}(y_{ij} - \bar{y}_j)^{2}  $$<br><br>
                        <br><br><br>
                </ul>
            <li> <b>Calcule a Média dos Quadrados (MS)</b>:
                <ul>
                    <li> Média dos Quadrados dos Grupos (MSG): 
                        $$MSG = \frac{SSG}{k-1} $$ <br>
                    <li> Média dos Quadrados dos Resíduos (MSR): 
                        $$MSG = \frac{SSR}{N-k} $$
                        <br><br>
                        Aqui,$k$ é o número de grupos e $N$ é o número total de observações
                        <br><br><br>
                </ul>
            <li> <b>Calcule a estatística $F$</b>:
                <br>
                $$F = \frac{MSG}{MSR} $$ <br>
            <li> <b>Observe o valor F na tabela F</b>: <br>
                O valor F é encontrado observando os graus de liberdade no numerador e o denominador na tabela F. <br><br><br>
            <li> <b>Compare a estatística $F$ obtida na etapa 4 com o valor crítico da etapa 5</b>: <br>
                Se a estatística $F$ for maior que o valor crítico no nível de significância exigido, rejeitamos a hipótese nula. <br><br><br>
        </ol>
    </div>

```python
group1 = car_clean[['drive-wheels', 'price']].groupby('drive-wheels')

f, p = scipy.stats.f_oneway(group1.get_group('fwd')['price'], group1.get_group('4wd')['price'], group1.get_group('rwd')['price'])
print(f'O valor do F-test = {f} e do P-value = {p}')
```

```python
for i in car_clean['drive-wheels'].unique():
    for k in car_clean['drive-wheels'].unique():
        if i < k:
            x, y = scipy.stats.f_oneway(group1.get_group(i)['price'], group1.get_group(k)['price'])
            print(f'Os values {i} e {k} possuem o F-test = {x} e o P-value = {y}')
```

<center><b><span style='font-size: 26px'>Tópicos em Machine Learning</span></center></b><br><br>
   Machine Learning, ou Aprendizado de Máquina, é um subcampo da Inteligência Artificial (IA) que se concentra no desenvolvimento de algoritmos e modelos estatísticos que permitem aos computadores aprender e fazer previsões ou decisões sem serem explicitamente programados para realizar a tarefa. <br><br>

Existem três tipos principais de aprendizado de máquina:<br>

<b>Aprendizado Supervisionado:</b> O modelo é treinado em um conjunto de dados que já contém a resposta desejada. O modelo aprende a prever a saída a partir dos dados de entrada.
<br><br>
<b>Aprendizado Não Supervisionado:</b> O modelo é treinado em um conjunto de dados sem respostas pré-definidas. O objetivo é encontrar padrões e estruturas nos dados.
<br><br>
<b>Aprendizado por Reforço:</b> O modelo aprende a tomar decisões com base em recompensas e punições. Ele é treinado para maximizar a recompensa ao longo do tempo.
<br><br>
Os algoritmos de aprendizado de máquina são usados em uma variedade de aplicações, desde sistemas de recomendação até carros autônomos.

<span style='font-size: 19px; font-weight:bold'>Tipos de aprendizados supervisionados</span><br><br><br>
   
<b>Regressão:</b> Este é um tipo de problema de aprendizado supervisionado onde o objetivo é prever valores contínuos. Exemplos incluem a previsão do preço de uma casa com base em características como o número de quartos, a idade da casa, a localização, etc.

<b>Classificação:</b> Este é um tipo de problema de aprendizado supervisionado onde o objetivo é prever uma classe ou categoria. Exemplos incluem a previsão de se um e-mail é spam ou não, ou se uma transação de cartão de crédito é fraudulenta ou não.

<b>Ranking:</b> Este é um tipo de problema de aprendizado supervisionado onde o objetivo é prever a ordem ou classificação de exemplos. Isso é comumente usado em sistemas de recomendação para prever a ordem dos itens que um usuário pode gostar.

<b>Detecção de anomalias:</b> Embora muitas vezes seja enquadrado como um problema de aprendizado não supervisionado, também pode ser tratado como um problema de aprendizado supervisionado onde o objetivo é prever se um exemplo é uma anomalia ou não com base em exemplos anteriores de comportamento normal e anômalo.

<b>Aprendizado de sequência:</b> Este é um tipo de problema de aprendizado supervisionado onde o objetivo é prever elementos em uma sequência. Exemplos incluem a previsão da próxima palavra em uma frase ou a próxima nota em uma peça musical. <br><br><br>

<span style='font-size: 19px; font-weight:bold'>Tipos de aprendizados não supervisionados</span>
<br><br><br>
<b>Clustering:</b> Este é um tipo de problema de aprendizado não supervisionado onde o objetivo é agrupar exemplos semelhantes juntos. Exemplos incluem a identificação de grupos de clientes com comportamentos de compra semelhantes.<br><br>

<b>Redução de dimensionalidade:</b> Este é um tipo de problema de aprendizado não supervisionado onde o objetivo é reduzir o número de recursos em um conjunto de dados enquanto mantém a maior parte da informação. Exemplos incluem a análise de componentes principais (PCA).<br><br>

<b>Detecção de anomalias:</b> Este é um tipo de problema de aprendizado não supervisionado onde o objetivo é identificar exemplos que são anômalos ou significativamente diferentes do resto. Exemplos incluem a detecção de fraudes de cartão de crédito.<br><br>

<b>Aprendizado de regras de associação:</b> Este é um tipo de problema de aprendizado não supervisionado onde o objetivo é descobrir relações interessantes entre variáveis em grandes bancos de dados. Um exemplo comum é a análise de cesta de mercado, que examina quais produtos tendem a ser comprados juntos.<br><br>

<b>Autoencoders:</b> Este é um tipo de rede neural usada para aprendizado não supervisionado. O objetivo é aprender uma representação compacta e eficiente dos dados de entrada, geralmente para o propósito de redução de dimensionalidade ou detecção de anomalias.<br><br><br>

<span style='font-size: 19px; font-weight:bold'>Tipos de aprendizados não supervisionados</span><br><br><br>
<b>Aprendizado por Reforço Baseado em Valor:</b> Este tipo de aprendizado por reforço se concentra em aprender uma função de valor que indica quão bom é estar em um determinado estado ou quão boa é uma determinada ação em um estado. O algoritmo mais conhecido nesta categoria é o Q-Learning.<br><br>

<b>Aprendizado por Reforço Baseado em Política:</b> Este tipo de aprendizado por reforço se concentra diretamente na aprendizagem da política, que é a estratégia que o agente usa para determinar a próxima ação com base no estado atual. <b>Aprendizado por Reforço Baseado em Modelo:</b> Este tipo de aprendizado por reforço envolve a criação de um modelo do ambiente, que o agente pode usar para planejar. Isso permite que o agente pense à frente e simule diferentes ações antes de tomá-las.

<b>Aprendizado por Reforço Profundo (Deep Reinforcement Learning):</b> Este é um subcampo do aprendizado por reforço que combina redes neurais profundas com aprendizado por reforço. O método <i style='font-weight: 500'>PolynomialFeatures(degree=)</i> da biblioteca scikit-learn é usado para gerar uma nova matriz de recursos que consiste em todas as combinações polinomiais dos recursos com grau menor ou igual ao grau especificado.<br><br>

Por exemplo, se uma amostra de entrada é bidimensional e da forma $[a, b]$, os recursos polinomiais de grau 2 são $[1, a, b, a^2, ab, b^2]$.

Os parâmetros do método são:<br>
    <ul>
   <li> <b>degree:</b> Grau máximo dos recursos polinomiais. Se uma tupla (min_degree, max_degree) for passada, então min_degree é o grau polinomial mínimo e max_degree é o grau polinomial máximo dos recursos gerados1.<br><br>
    <li><b>interaction_only:</b> Se verdadeiro, apenas os recursos de interação são produzidos: recursos que são produtos de no máximo degree recursos de entrada distintos1. <br><br>
    <li><b>include_bias:</b> Se verdadeiro (padrão), inclui uma coluna de viés, o recurso em que todos os poderes polinomiais são zero (ou seja, uma coluna de uns - atua como um termo de interceptação em um modelo linear).
    Este método é útil para modelar efeitos não lineares com um modelo linear, adicionando recursos não lineares. No entanto, esteja ciente de que o número de recursos na matriz de saída aumenta polinomialmente com o número de recursos da matriz de entrada e exponencialmente com o grau. Graus altos podem causar sobreajuste.
    </ul>
    <br><br>
    Os métodos <b>fit</b> e <b>fit_transform</b> são comumente usados em módulos de pré-processamento de dados do scikit-learn, como PolynomialFeatures.<br>

<b>fit:</b> Este método é usado para calcular os parâmetros necessários para a transformação. Por exemplo, no caso do StandardScaler, o método fit calcula a média e o desvio padrão dos recursos. Esses parâmetros são então usados para transformar os dados.<br>

<b>fit_transform:</b> Este método combina as funcionalidades de fit e transform em uma única etapa. Primeiro, ele ajusta o modelo aos dados e depois transforma os dados de acordo com o modelo ajustado1. Isso pode ser mais eficiente do que chamar fit e transform separadamente.<br>

Por exemplo, se você estiver usando o PolynomialFeatures(degree=2), o método fit calculará os parâmetros necessários para transformar seus recursos em todas as combinações polinomiais de grau 2. Em seguida, transform aplicará essa transformação aos seus dados. Se você usar fit_transform, ambos os passos serão realizados em uma única chamada. <br>

Lembre-se de que a transformação é aplicada aos dados de treinamento e também deve ser aplicada aos dados de teste. No entanto, você deve ajustar o transformador apenas nos dados de treinamento para evitar vazamento de informações dos dados de teste. <br>

<b><span style='font-size: 20px'>Pipelines</span></b><br><br><br>

Um **data pipeline** é uma estrutura que permite a coleta, transformação e movimentação de dados, desde sua origem até seu destino final. Ele desempenha um papel crucial na preparação dos dados para a análise, garantindo que eles sejam processados, limpos e estruturados corretamente.<br>

O data pipeline é composto por etapas que envolvem a extração, transformação e carregamento de dados (ETL, na sigla em inglês). A primeira etapa é a extração, na qual os dados são coletados de diferentes fontes, como bancos de dados, APIs, arquivos ou sistemas externos. Em seguida, ocorre a etapa de transformação, na qual os dados são processados, limpos e preparados para análise. Essa etapa pode envolver a padronização de formatos, a correção de erros, a agregação de dados e a aplicação de regras de negócio. Por fim, os dados transformados são carregados no destino final, como um data warehouse, um banco de dados ou uma plataforma de análise.<br>

A essência do data pipeline está em criar um fluxo contínuo e automatizado de dados, garantindo que as informações necessárias estejam disponíveis para análise de forma eficiente e confiável. Ao estabelecer um pipeline robusto, é possível evitar inconsistências, duplicações ou erros nos dados, o que impactaria diretamente a qualidade e a precisão dos insights gerados pela análise.<br>

Um data pipeline bem projetado permite que as organizações obtenham benefícios significativos na análise de dados. Ele oferece agilidade ao permitir que os dados sejam atualizados e processados em tempo real ou com frequência definida. Além disso, um pipeline eficiente facilita a manutenção e o gerenciamento dos dados, permitindo que as equipes de análise se concentrem no que realmente importa: a extração de insights valiosos.<br>

Outro aspecto importante é a escalabilidade do data pipeline. Com a explosão de dados nos dias de hoje, é essencial ter uma infraestrutura capaz de lidar com grandes volumes de informações. O data pipeline pode ser dimensionado para lidar com um aumento no volume de dados, garantindo que a análise não seja limitada pela capacidade de processamento.<br>

<b><span style='font-size: 20px'>Measures for in-Sample Evaluation</span></b><br><br><br>

Uma maneira de avaliar a qualidade de um modelo de regressão é usando métricas como <b>Coeficiente de Determinação $R^2$</b> e o <b>Erro Quadrático Médio</b> (Mean Squared Error - MSE).
<br><br>
<ul>
    <li><b>Coeficiente de Determinação - $R^2$</b>:<br>
    O coeficiente de determinação, também chamado de $R^2$, é uma medida de ajuste de um modelo estatístico linear generalizado, como a regressão linear simples ou múltipla, aos valores observados de uma variável aleatória. O $R^2$ varia entre 0 e 1, por vezes sendo expresso em termos percentuais. Ele expressa a quantidade da variância dos dados que é explicada pelo modelo linear. Assim, quanto maior o $R^2$, mais explicativo é o modelo linear, ou seja, melhor ele se ajusta à amostra.
$$ R^2 = 1 - \frac{SSR}{SST}$$
<br> Em que $SSR$ e $SST$ são a soma dos quadrados dos resíduos e soma total dos quadrados respectivamente.<br><br>
<li><b>Mean Squared Error - </b>$MSE$:<br>
   O erro quadrático médio - $MSE$, é comumente usado para verificar a acurácia de modelos e dá um maior peso aos maiores erros, já que, ao ser calculado, cada erro é elevado ao quadrado individualmente e, após isso, a média desses erros quadráticos é calculada.
    $$ MSE = \frac{1}{n} \sum_{i=1}^{n}(y_i - \hat{y}_i)^2 $$
    onde $n$ é o número de amostra, $y_i$ é o valor observado para a amostra i e $\hat{y}_i$ é o valor previsto pelo modelo para a amostra $i$.
</ul><br><br><br>

Para calcular $R^2$, podemos fazer isso de duas formas: Usando o médoto <i style='font-weight: 500'>lr.score(X,Y)</i> do módulo sklearn.linear_model quando usado uma regressão linear ou usando o método <i style='font-weight: 500'>r2_score($Y$,$\hat{Y}$)</i> do módulo sklearn.metrics. Este segundo também tem uma função para calcular o $MSE$. Para fazer isso, basta usar o método <i style='font-weight: 500'>mean_squared_error($Y$,$\hat{Y}$)</i>

```python
from sklearn import linear_model
lr = sklr.linear_model.LinearRegression()
lr.fit(car_clean[['highway-L/100km']],car_clean['price'])
lr.score(car_clean[['highway-L/100km']],car_clean['price'])
lr1 = sklr.linear_model.LinearRegression()
lr1.fit(car_clean[['highway-L/100km', 'horsepower', 'curb-weight', 'engine-size']], car_clean['price'])
```
