```python

changer = {'two': int(2),'three':int(3), 'four':int(4),'five':int(5),'six': int(6),'eight': int(8),'twelve': int(12), '?':np.nan }
cardfpre['num-of-doors'] = cardfpre['num-of-doors'].apply(lambda x: changer[x] )
cardfpre['num-of-cylinders'] = cardfpre['num-of-cylinders'].apply(lambda x: changer[x] )
cardfpre
```

Agora basta transformar os objects em numeric

```python
for coluna in cardfpre[colunas_conv]:
    cardfpre[coluna] = pd.to_numeric(cardfpre[coluna], errors='coerce')
cardfpre.info()
```

Nestes dados, precisamos decidir o que fazer com valores nulos. Note que devemos fazer nossas escolhas baseado na relevância de cada coluna para nossas análises futuras.


```python
cardfpre.isna().any()
```

As colunas normalized-losses, num-of-doors, bore, stroke, horsepower, peak-rpm, price contém valores Nulos, decidiremos o que fazer uma por uma.<br>
<ul><li>Para as colunas normalized-losses, stroke, bore, horsepower and peak-rpm iremos usar a média de cada coluna.</li>
    <li>Para a coluna 'num-of-doors' iremos usar a frequência <span style='font-style:italic'>DataFrame[column].value_counts()</span></li>
    <li>Para a coluna price iremos deletar as linhas já que precisamos desse valor para nossas predições</li>

</ul>

```python
media_n_l = cardfpre['normalized-losses'].mean()
cardfpre['normalized-losses'].replace(np.nan, media_n_l, inplace=True)
cardfpre['normalized-losses'].isna().any()
 
```

Agora faremos para todas as outras colunas

```python
transform = ['bore', 'stroke',  'horsepower', 'peak-rpm']
cardfpre[transform]
```

```python
media = 0
for coluna in cardfpre[transform]:
    media = cardfpre[coluna].mean()
    cardfpre[coluna].replace(np.nan, media, inplace=True)
cardfpre.isna().any()
```

```python
doors_null = cardfpre['num-of-doors'].isna()
cardfpre[doors_null]
```

Os dois valores NaN são de sedans, faremos então uma busca de todos os sedans e veremos a frequência do número de portas

```python
sedans = cardfpre['body-style'] == 'sedan'
cardfpre[sedans]
```

```python
cardfpre[sedans]['num-of-doors'].value_counts().to_frame()
```

```python
print(f'A quantidade de sedans de quatro portas corresponde a {(79 / cardfpre[sedans]['num-of-doors'].shape[0]):.3f}% do total de sedans')
```

Desse modo, substituiremos os valores NaN por 4.0


```python
moda = cardfpre[sedans]['num-of-doors'].mode()
cardfpre['num-of-doors'].replace(np.nan, 4.0, inplace=True)
cardfpre['num-of-doors'].isna().any()
```
