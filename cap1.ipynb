Now Let's define $\Omega$ as $\Omega = \{(\omega_i, \omega_j) \; | \;\omega_i \neq \omega_j, \; \omega_i, \omega_j \in S\} $ the set of all different ways to draw two balls without replacement and $\mathcal{A}(\Omega)$ a sigma-algebra for it. This means $|\Omega| = (n + k)\times (n+k -1)$<br><br><br><br>
<ul> 
    <li><b>$(r_i,r_j)$:</b> &nbsp; &nbsp;We could describe the set where the first ball drawn is red as $A = \{(r_i, \omega_j) \; | \; r_i \neq \omega_j, \;\; r_i, \omega_j \in S\} $ and the set where the second ball is red as $B = \{(\omega_i, r_j) \; | \: \omega_i \neq r_j, \; \;  \omega_i, r_j \in S\}$ and $A \cap B = \{(r_i, r_j) \; | \; r_i \neq r_j, \; \; r_i, r_j \in S\}$ as the set where both balls drawn are red. Note that $|A \cap B| = n\times (n-1)$<br>
        Then the probability of picking two red balls is  &nbsp; 
        $$P(A \cap B) = \frac{n \times (n-1)}{(n + k)\times (n+k -1)}$$.     <br><br>                                                                                                                                                                                            
    <li><b>$(r_i,b_j)$:</b>&nbsp; &nbsp; let's the set $C = \{(\omega_i, b_j) \; | \: \omega_i \neq b_j, \; \;  \omega_i, b_j \in S\}$ be the set where the second ball drawn is red. Hence $A \cap C = \{(r_i, b_j) \; | \;\; \; r_i, b_j \in S\}$ and $|A \cap C| = n \times k$ and finally
    $$P(A \cap C) = \frac{n\times k }{(n + k)(n + k -1)} = \frac{1}{n + k + 1}$$ <br><br>


<li>$(b_i,r_j)$:</b>&nbsp; &nbsp; Define $D= \{(b_i,\omega_j) \; | \:b_i \neq  \omega_i, \; \; b_i, \omega_j\in S\} $ as the set where the first ball drawn is black. Since $D \cap B = \{(b_i, r_j) \; | \;\; \; b_i, r_j \in S\} \rightarrow |D \cap B| = n \times k$ is known then 
$$P(D \cap B) =\frac{1}{n + k + 1} $$<br><br>

<li>$(b_i,b_j)$:</b>&nbsp; &nbsp; $C \cap D = \{(b_i, b_j) \; | \; b_i \neq b_j, \; \; b_i, b_j \in S\} \rightarrow |C \cap D| = k \times (k-1)$. Consequently 
$$P(C \cap D) = \frac{k \times (k-1)}{(n + k)\times (n+k -1)} $$
    
        
</ul>

If a red ball is drawn, then $S$ keep the same. Otherwise if a black ball was drawn then $S \rightarrow S_1 = \{r_1, ..., r_{10}, b_1, ..., b_5, b_6, b_7\}$. Thus
$$\Omega = \{(\omega_i, \omega_j) \; | \; \omega_i, \omega_j \in S \times S \;\; {\text{OR}} \;\; \omega_i, \omega_j \in S \times S_1 \} $$
Define $A_1$ as the set where the first ball drawn is red and $B_1$ as the set where the first ball drawn is black. Then $A_1 = \{(r_i, \omega_j) \; | \; \;\; r_i, \omega_j \in S\} \rightarrow P(A_1) = \frac{10 \times 15}{15 \times 15} = \frac{2}{3}$ &nbsp; &nbsp; and $B_1 = \{(b_i, \omega_j) \; | \; \;\; b_i, \omega_j \in S\} \rightarrow P(B_1) = \frac{5 \times 17}{15 \times 17} = \frac{1}{3}$. 

<br><br>

<ul>

        $$P(B_2) = P(B_2 | A_1) \times P(A_1) + P(B_2 | B_1) \times P(B_1) = \frac{1}{3} \times \frac{2}{3} + \frac{7}{17} \times \frac{1}{3} = \frac{51}{153} $$
 </ul>

    <li> Let $B_1 = \{(w_i,w_j) \;\; | \;  B_1 \in (\Omega)\} $ be the set where two white balls were drawn and $B_2 = \{(b_i,b_j) \;\; | \;  B_2 \in \Omega\} $ the set where two black balls were drawn. Therefore $$|B_1| = 3^2 \;\; \rightarrow \;\; P(B) = \frac{|B|}{|\Omega|} = \frac{9}{25}$$
        $$|B_2| = 2^2 \;\; \rightarrow \;\; P(B) = \frac{|B|}{|\Omega|} = \frac{4}{25}$$
        $$ P(B) = \frac{13}{25}$$ <br><br>
    <li> Let $C = \{(\omega_i, \omega_j) \;\; | \;\; $ at least one of $\omega_{i,j} \in \{w_1, w_2, w_3\} \}$. Another way to see this problem is to note that $\complement C = \{(b_i, b_j) \;\; | \;\; \complement{C} \in \Omega\} \;\; \rightarrow \;\; | \complement C | = 4$. Therefore
        $$P(C) = 1 - P(\complement{C}) = 1 - \frac{4}{25} = \frac{21}{25}$$
</ul>
